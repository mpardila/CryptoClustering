```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
market_scaled = StandardScaler().fit_transform(df_market_data)
```

```python
# Create a DataFrame with the scaled data
market_scaled_df = pd.DataFrame(market_scaled, columns=df_market_data.columns)

# Copy the crypto names from the original data
market_scaled_df["coin_id"] = df_market_data.index

# Set the coinid column as index
market_scaled_df = market_scaled_df.set_index("coin_id")

# Display sample data
market_scaled_df.head()
```

```python
# Plot your scaled data to see what's in your DataFrame
market_scaled_df.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k_scaled = list(range(1,12))
```

```python
# Create an empty list to store the inertia values
inertia_scaled = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for i in k_scaled:
    k_model_scaled = KMeans(n_clusters=i, random_state=1)
    k_model_scaled.fit(market_scaled_df)
    inertia_scaled.append(k_model_scaled.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_scaled_data = {"k": k_scaled, "inertia": inertia_scaled}

# Create a DataFrame with the data to plot the Elbow curve
elbow_scaled_df = pd.DataFrame(elbow_scaled_data)
elbow_scaled_df.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_scaled = elbow_scaled_df.hvplot.line(
                                x="k",
                                y="inertia",
                                title="Elbow Scaled Curve",
                                xticks=k_scaled
                                )
elbow_scaled
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** Based on the elbow graph of the scaled data, the best value of k is 4

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
model_scaled = KMeans(n_clusters=4, random_state=1)
model_scaled
```

```python
# Fit the K-Means model using the scaled data
model_scaled.fit(market_scaled_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
scaled_clusters = model_scaled.predict(market_scaled_df)

# Print the resulting array of cluster values.
print(scaled_clusters)
```

```python
# Create a copy of the DataFrame
market_scaled_predictions_df = market_scaled_df.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
market_scaled_predictions_df["scaled_cluster"] = scaled_clusters

# Display sample data
market_scaled_predictions_df.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
scaled_scatter = market_scaled_predictions_df.hvplot.scatter(
                                                        x="price_change_percentage_24h",
                                                        y="price_change_percentage_7d",
                                                        by="scaled_cluster",
                                                        hover_cols="coin_id",
                                                        title="Scaled Clusters"
                                                        )
scaled_scatter
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
market_pca = pca.fit_transform(market_scaled_df)

# View the first five rows of the DataFrame. 
market_pca[:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca.explained_variance_ratio_
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** The first principal component explains approximately 37.20% of the variance of the data. The second principal component explains approximately 34.70% of the variance. The third principal component explains only 17.60% of the variance. The three components together capture 89.50% of the informatuon in the original dataset, leaving 10.5% of the cariance inaccounted for.

```python
# Create a new DataFrame with the PCA data.
market_pca_df = pd.DataFrame(
    market_pca,
    columns=["PCA1", "PCA2", "PCA3"]
)

# Copy the crypto names from the original data
market_pca_df["coin_id"] = df_market_data.index

# Set the coinid column as index
market_pca_df = market_pca_df.set_index("coin_id")

# Display sample data
market_pca_df.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k_pca = list(range(1,12))
```

```python
# Create an empty list to store the inertia values
inertia_pca = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for i in k_pca:
    k_model_pca = KMeans(n_clusters=i, random_state=1)
    k_model_pca.fit(market_pca_df)
    inertia_pca.append(k_model_pca.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_pca_data = {"k": k_pca, "inertia": inertia_pca}

# Create a DataFrame with the data to plot the Elbow curve
elbow_pca_df = pd.DataFrame(elbow_pca_data)
elbow_pca_df.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_pca = elbow_pca_df.hvplot.line(
                                x="k",
                                y="inertia",
                                title="Elbow Curve PCA",
                                xticks=k_pca
)
elbow_pca
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** Based on the elbow graph of the PCA data, the best value of k is 4


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No, the best value of k in both scenarios is the same

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
model_pca = KMeans(n_clusters=4, random_state=1)
model_pca
```

```python
# Fit the K-Means model using the PCA data
model_pca.fit(market_pca_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
pca_clusters = model_pca.predict(market_pca_df)

# Print the resulting array of cluster values.
print(pca_clusters)
```

```python
# Create a copy of the DataFrame with the PCA data
market_pca_predictions_df = market_pca_df.copy()

# Add a new column to the DataFrame with the predicted clusters
market_pca_predictions_df["pca_cluster"] = pca_clusters

# Display sample data
market_pca_predictions_df.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
pca_scatter = market_pca_predictions_df.hvplot.scatter(
                                                x="PCA1",
                                                y="PCA2",
                                                by="pca_cluster",
                                                hover_cols="coin_id",
                                                title="PCA Clusters"
                                                )
pca_scatter
```
